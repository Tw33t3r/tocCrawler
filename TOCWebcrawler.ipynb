```python
#setup and get htmlDirectory
import requests

url = "http://www.wuxiaworld.com/novel/wu-dong-qian-kun"
htmlDirectory = requests.get(url)
```

```python
#save to a file
directory = open('directory.txt', 'w')
directory.write(htmlDirectory.text)
directory.close()
```

```python
#find relevant information from html, save to a list
chapters = []
bookLink = "/novel/wu-dong-qian-kun/"
directory = open('directory.txt', 'r')
for line in directory:
    if ((line.find(bookLink))!= -1):
        _,_,line = line.rpartition('/')
        line = line[:-3]
        chapter = url + '/' + line
        chapters.append(chapter)
directory.close()
```

```python
#for counter, chapterURL in enumerate(chapters):
chapterHTML = requests.get(chapters[0])
#chapterString = "chapter" + str(counter + 1) + ".html"
chapterFile = open('chapter0.txt', 'w')
chapterFile.write(chapterHTML.text)
chapterFile.close()
#find content of chapter
chapterFile = open('chapter0.txt', 'r')
content = str()
countdown = -1
for line in chapterFile:
    countdown = countdown - 1
    if ((line.find('h4 class'))!= -1):
        countdown = 4
    if (countdown == 0):
        content = line
        break
    print(countdown)
chapterFile.close()
finalHTML = open('final.html', 'w')
finalHTML.write(content)
```
